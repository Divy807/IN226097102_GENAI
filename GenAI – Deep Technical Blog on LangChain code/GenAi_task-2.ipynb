LLMs and Chat Models

```python
pipe = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100
)

llm = HuggingFacePipeline(pipeline=pipe)

response = llm.invoke("Explain neural networks in simple terms")
print(response)
```

Prompts and Prompt Templates

template = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

prompt = template.format(topic="LangChain")
print(prompt)

Chains

```python
chain = template | llm

result = chain.invoke({"topic": "Machine Learning"})
print(result)
```

Memory

```python
history = []

def conversation(user_input):
    history.append(f"User: {user_input}")
    
    prompt = "\n".join(history) + "\nAI:"
    
    response = llm.invoke(prompt)
    history.append(f"AI: {response}")
    
    return response

print(conversation("Hi, I am Divya"))
print(conversation("What is my name?"))
```

Agents

```python
def calculator_tool(input_text):
    return eval(input_text)

query = "25 * 4"

if any(op in query for op in ["*", "+", "-", "/"]):
    result = calculator_tool(query)
else:
    result = llm.invoke(query)

print(result)
```
